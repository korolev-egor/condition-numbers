```python
def Jacobi_method(a, epsilon, method = 1):  # Нахождение собственных чисел методом Якоби
    def t(i, j):  # Строим матрицу вращения
        x = -2*a[i, j]
        y = a[i, i] - a[j, j]
        if y == 0:
            cos = sin = 1/np.sqrt(2)
        else:
            cos = np.sqrt((1+abs(y)/np.sqrt(x**2+y**2))/2)
            sin = np.sign(x*y)*abs(x)/(2*cos*np.sqrt(x**2+y**2))
        m = matrix(np.eye(n, n))
        m[i, i] = cos
        m[i, j] = -sin
        m[j, j] = cos
        m[j, i] = sin
        return m
    
    def r(i):  # с
        return sum([abs(a[i, j]) for j in range(n) if j != i])
    
    def search_max():
        res = [0]
        for i in range(n):
            for j in range(i+1, n):
                if abs(a[i, j]) > abs(res[0]):
                    res = a[i, j], i, j
        return res[1], res[2]
    
    def cycle():
        nonlocal max_
        if max_[0] == n-2:
            max_ = (0, 1)
        elif max_[1] == n-1:
            max_ = (max_[0]+1, max_[0]+2)
        else:
            max_ = (max_[0], max_[1]+1)
    
    iteration = 0
    condition = True
    n = a.shape[0]
    max_ = [0, 0]
    while condition:
        iteration += 1
        if method == 1:
            max_ = search_max()
        else:
            cycle()
        turn = t(*max_)
        a = turn*a*turn.T
        condition = sum([r(i)>epsilon for i in range(n)])
    eigens = []
    for i in range(n):
        eigens.append(a[i, i])
    return sorted(eigens), iteration
```
